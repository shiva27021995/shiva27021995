```python
crime = crime.astype({"Latitude": "object", "Logitude": "object"})
```

```python
crime.isnull().sum()
```

```python
# Assign X (data) and y (target)


X = crime.iloc[:,15:18]
y = crime["CommunityArea"]
print(X.shape, y.shape)
```

```python
X
```

```python
y
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
#strafity helps to make the data more random

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.10, random_state=2)
```

```python
#create a logisitc regression model 
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', max_iter=100000)
classifier
```

```python
#fit (train) model with training data 
classifier.fit(X_train, y_train)
```

```python
#validata model using test data
#
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
#make predictions 
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```

# RandomForest Algorithm

```python
#### random forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)

```

```python
print(f"Training Data Score: {rf.score(X_train, y_train)}")
print(f"Testing Data Score: {rf.score(X_test, y_test)}")
```

```python
predictions = rf.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```
